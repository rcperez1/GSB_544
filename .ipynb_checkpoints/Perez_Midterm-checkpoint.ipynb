```python
def recommend_trails(midterm, area_name, min_elev=None, max_elev=None, min_rating=None, route_type=None):
    
    if min_elev is None:
        min_elev = min(midterm['elevation_gain'] for trail in dataset if trail['area_name'] == area_name)

    recommended_trails = []

    for trail in midterm:
        if midterm['area_name'] == area_name:
            if min_elev is not None and trail['elevation_gain'] < min_elev:
                continue
            if max_elev is not None and trail['elevation_gain'] > max_elev:
                continue
            if min_rating is not None and trail['avg_rating'] < min_rating:
                continue
            if trail_type is not None and trail['route_type'].lower() != trail_type.lower():
                continue
            recommended_trails.append(trail)

    return recommended_trails
```
